```python
import torch
torch.manual_seed(512)
```

```python
# 定义线性模型和sigmoid
class Linear:
  # input:(B, in_features)
  # output:(B, out_features)
  def __init__(self, in_features, out_features, bias = True):
    # 对于模型参数初始化，没有优化。
    self.weight = torch.randn(in_features, out_features, requires_grad = True)
    if bias:
      self.bias = torch.randn(out_features, requires_grad = True)
    else:
      self.bias = None

  def __call__(self, x):
    # x: (B, in_features)
    # self.weight: (in_features, out_features)
    self.out = x @ self.weight
    if self.bias is not None:
      self.out += self.bias
    return self.out

  def parameters(self):
    # 返回模型参数
    if self.bias is not None:
      return [self.weight, self.bias]
    else:
      return [self.weight]
```

```python
l = Linear(3, 4)
x = torch.randn(5, 3)
l(x).shape
```

```python
class Sigmoid:
  def __call__(self, x):
    self.out = torch.sigmoid(x)
    return self.out

  def paramaters(self):
    return []
```

```python
s = Sigmoid()
x = torch.randn(3, 2)
print("x:\n", x)
s(x).shape
print("s(x):\n", s(x))
```
