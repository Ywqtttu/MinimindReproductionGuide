```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import string

# 定义字典
char2idx = {s: i for i, s in enumerate(string.ascii_lowercase)}
```

```python
eg = "love"
idx = []
for i in eg:
  idx.append(char2idx[i])
idx = torch.tensor(idx)
idx
```

```python
# 使用onehot编码，将文本转换为二维张量
num_claz = len(char2idx.keys())
x = F.one_hot(idx, num_classes = num_claz).float()
x, x.shape
```

```python
dims = 5
w = torch.randn(num_claz, dims)
(x @ w), (x @ w).shape
```

```python
w[idx], w[idx].shape, w.shape, idx
# 和上文的结果一致，故而不会多做上面这一步热独编码
```

```python
class Embedding:

  def __init__(self, num_embeddings, embedding_dim):
    self.weight = torch.randn((num_embeddings, embedding_dim), requires_grad = True)

  def __call__(self, x):
    self.out = self.weight[x]
    return self.out

  def paramters(self):
    return [self.weight]
```

```python
em = Embedding(num_claz, 5)
x = torch.randint(0, num_claz, (10, ))
em(x).shape
```
